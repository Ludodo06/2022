```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
nb_deaths    = 33000
nb_suicides  = 2/3 * nb_deaths
nb_homocides = 1/3 * nb_deaths

nb_suicides_male   = 0.85 * nb_suicides
nb_suicides_female = 0.15 * nb_suicides

nb_homocides_female = 0.15 * nb_homocides
nb_homocides_male_young_black = 0.5 * 1/3 * nb_homocides
```

```python
list_suicides  = np.arange(0,nb_suicides)
list_homocides = np.arange(0,nb_homocides)

plt.figure(figsize=(14.01,4.25))
plt.plot((list_suicides/100).astype(int),list_suicides%100,'ro',markersize=0.8)
plt.plot(((list_homocides+nb_suicides)/100).astype(int),(list_homocides+nb_suicides)%100,'go',markersize=0.8)
plt.show()
```

```python
list_suicides_male   = np.arange(0,nb_suicides_male)
list_suicides_female = np.arange(0,nb_suicides_female)

plt.figure(figsize=(14.01,4.25))
plt.plot((list_suicides_male/100).astype(int),list_suicides_male%100,'o',markersize=0.8,color='lightcoral')
plt.plot(((list_suicides_female+nb_suicides_male)/100).astype(int),(list_suicides_female+nb_suicides_male)%100,'o',markersize=0.8,color='indianred')
plt.plot(((list_homocides+nb_suicides_female+nb_suicides_male)/100).astype(int),(list_homocides+nb_suicides_female+nb_suicides_male)%100,'o',markersize=0.8,color='green')
plt.show()
```

```python
list_homocides_female           = np.arange(0,nb_homocides_female)
list_homocides_male_young_black = np.arange(0,nb_homocides_male_young_black)
list_homocides_others           = np.arange(0,nb_homocides-nb_homocides_female-nb_homocides_male_young_black)

plt.figure(figsize=(14.01,4.25))
plt.plot((list_suicides/100).astype(int),list_suicides%100,'ro',markersize=0.8)
plt.plot(((list_homocides_female+nb_suicides)/100).astype(int),(list_homocides_female+nb_suicides)%100,'o',markersize=0.8,color='forestgreen')
plt.plot(((list_homocides_male_young_black+nb_homocides_female+nb_suicides)/100).astype(int),(list_homocides_male_young_black+nb_homocides_female+nb_suicides)%100,'o',markersize=0.8,color='limegreen')
plt.plot(((list_homocides_others+nb_homocides_female+nb_suicides+nb_homocides_male_young_black)/100).astype(int),((list_homocides_others+nb_homocides_female+nb_suicides+nb_homocides_male_young_black)%100).astype(int),'o',markersize=0.8,color='green')
plt.show()
```
